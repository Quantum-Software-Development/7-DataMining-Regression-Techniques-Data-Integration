```python
# Import necessary libraries for Max-Min Normalization
import pandas as pd
from sklearn.preprocessing import minmax_scale
```

```python
# Load the data from the CSV file
df = pd.read_csv('/content/cancer.csv')

# Get the column names of the DataFrame as a list.
# Then remove 'id' and 'diagnosis' from the list 'cols' as we don't want to apply minmax scale to these columns.
cols = list(df.columns)
cols.remove('diagnosis') # Column where we don't want to apply minmax scale

# Copy the data and apply minmax normalization to the columns with continuous values.
# By default, the minmax_scale method rescales with min=0 and max=1.
df_normalized = df.copy()
df_normalized[cols] = df[cols].apply(minmax_scale)

# Display the head of the normalized DataFrame
display(df_normalized.head())
```

```python
# Import necessary library for Z-score Standardization
from sklearn.preprocessing import StandardScaler

# Create a StandardScaler object
scaler = StandardScaler()

# Apply Z-score standardization to the selected columns
df_z_scaled = df[cols].copy() # Use the same 'cols' list as in Max-Min normalization
df_z_scaled = scaler.fit_transform(df_z_scaled)

# Convert the scaled data back to a DataFrame
df_z_scaled = pd.DataFrame(df_z_scaled, columns=cols)

# Display the head of the Z-score standardized DataFrame
display(df_z_scaled.head())
```

## Z-score Standardization Explanation

**English:**

Z-score standardization (or normalization) is a technique used to transform data so that it has a mean of 0 and a standard deviation of 1. It is calculated by subtracting the mean of the feature and then dividing by the standard deviation. This method is useful when the data has outliers or when the algorithm assumes a Gaussian distribution.

**Portuguese:**

A padronização Z-score (ou normalização) é uma técnica utilizada para transformar dados de forma que tenham média 0 e desvio padrão 1. É calculada subtraindo a média da característica e dividindo pelo desvio padrão. Este método é útil quando os dados possuem outliers ou quando o algoritmo assume uma distribuição Gaussiana.

```python
# Import necessary libraries for Z-score Standardization
import pandas as pd
from sklearn.preprocessing import scale

# Load the data from the CSV file (assuming it's needed for this cell)
df = pd.read_csv('/content/cancer.csv')

# Getting the column names of the DataFrame as a list.
cols = list(df.columns)

# Removing 'id' and 'diagnosis' from the list 'cols'
# cols.remove("id") # Keeping this commented out as 'id' column caused errors before
cols.remove("diagnosis")

# Copying the data and applying standardization to all columns of the DataFrame.
# By default, the scale method subtracts the mean and divides by the standard deviation.
df_z_scaled_another = df.copy() # Using a new variable name
df_z_scaled_another[cols] = df[cols].apply(scale)

# Display the standardized DataFrame
display(df_z_scaled_another.head())
```
